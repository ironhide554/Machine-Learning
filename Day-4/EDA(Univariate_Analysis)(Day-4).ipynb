```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df.head()
```

# Categorical Data

## Countplot
It gives us the frequency count of particular attributes.

```python
sns.countplot(x=df["Survived"])
df["Survived"].value_counts()
```

```python
sns.countplot(x=df["Survived"],  hue = df["Pclass"])
```

```python
sns.countplot(x = df["Pclass"])
```

```python
sns.countplot(x = df["Embarked"])
```

## Piechart

```python
df["Survived"].value_counts().plot(kind = "pie" , autopct = "%.2f")
```

# Numerical Variables

## Histogram
To understand the data in that numerical column.


```python
plt.hist(df["Age"])
```

```python
plt.hist(df["Age"] , bins = 50)
```

## Distplot
It is similar to histogram but contains (KDE) Kernel Density Estimation this plot provides us the probabilty of the values present at X-axis and it is a probability distribution function.

```python
sns.distplot(df['Age'])
```

## BoxPlot
it used for getting the 5 number summary of the column and it used for cheking the noisy data and for identifying the outliers.

```python
sns.boxplot(x = df["Fare"])
```

```python
sns.boxplot(x = df["Age"])
```

```python
df["Age"].skew() # Used for checking the skewness of the data 
```
